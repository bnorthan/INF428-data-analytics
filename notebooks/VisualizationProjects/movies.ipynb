```python
import pandas as pd
```

```python
movies=pd.read_csv('../../data/entertainment/film.csv', sep=';')
```

```python
movies.head()
```

```python
print movies.Year.min()
print movies.Year.max()
```

```python
# create groups
groups=movies.groupby('Actor')

# aggregate groups by count
actors = groups.Actor.count()

print actors.sort_values(ascending=False).head()
```

```python
# aggregate using mean
means=groups.mean()
print means.fillna(0).sort_values('Popularity',ascending=False).head()
```

```python
%matplotlib inline
means.Popularity.plot()
```
