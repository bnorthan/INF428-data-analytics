## Import and load data

Now import the pandas library and read the contents of the excel file into a variable called 'southpark'

```python
import pandas as pd
southpark=pd.read_excel('../../data/entertainment/all-seasons-small.xls')
```

Print out the type of the *southpark* variable (it should be a *dataframe*) and print the shape (using the *shape* property)

```python
type(southpark)
print southpark.shape
```

Now take a look at the *columns* property.  This shows us the *column* names

```python
print southpark.columns
```

Column names are important, as we can use them to directly access the columns.  For example below we print out the first 10 values in the *Character* column...

```python
print southpark.Character[0:10]
```

Let's get rid of the *Line* column (for now)

```python
southpark=southpark.drop("Line",1)
print southpark.columns
print southpark.shape
```

## Groupby
Group by character

```python
grouped=southpark.groupby('Character')
```

```python
type(grouped)
```

```python
print type(grouped.groups)
print type(grouped.get_group('Stan'))
print grouped.get_group('Stan')
```

Count the groups and sort 

```python
print type(grouped.count())
#print grouped.count()
sorted=grouped.count().sort_values('Season', ascending=False)
type(sorted)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.figure(figsize=(20,10))
sorted[0:10]['Season'].plot.bar()
```

```python
plt.figure(figsize=(20,10))
sorted[0:20]['Season'].plot.pie()
```
